#### 🎓 REDACCIÓN DEL PROBLEMA:
El rectángulo es una de las figuras geométricas más comunes y útiles en nuestra
vida diaria. Lo encontramos en pantallas de televisión, ventanas, puertas,
libros, mesas y countless objetos que nos rodean. Comprender cómo calcular sus
propiedades básicas como el área y el perímetro es fundamental para estudiantes
de preparatoria, ya que estas habilidades se aplican en arquitectura, diseño,
construcción y muchas otras áreas.

Un rectángulo es un cuadrilátero (figura de cuatro lados) con las siguientes
características:
1. Tiene cuatro lados
2. Sus lados opuestos son paralelos e iguales
3. Sus cuatro ángulos internos son rectos (90 grados)

Para definir un rectángulo, necesitamos conocer dos medidas:
1. La base (b) - el lado horizontal
2. La altura (h) - el lado vertical

El estudio del rectángulo nos permite resolver problemas prácticos como:
1. Calcular cuánta pintura se necesita para pintar una pared
2. Determinar cuántos metros de cerco se requieren para un terreno
3. Calcular el espacio disponible en una habitación..

1️⃣  ÁREA DEL RECTÁNGULO:
    ─────────────────────
    A = b × h
    
    Donde:
    • A = Área del rectángulo
    • b = base del rectángulo
    • h = altura del rectángulo
    
    El área representa la superficie interior del rectángulo y se mide en
    unidades cuadradas (cm², m², km², etc.)

2️⃣  PERÍMETRO DEL RECTÁNGULO:
    ─────────────────────────
    P = 2 × (b + h)
    
    Donde:
    • P = Perímetro del rectángulo
    • b = base del rectángulo
    • h = altura del rectángulo
    
    El perímetro representa la distancia total alrededor del rectángulo
    (la suma de todos sus lados) y se mide en unidades lineales (cm, m, km).

    También se puede expresar como:
    P = b + h + b + h = 2b + 2h

```python
# ==================== FUNCIONES DEL PROGRAMA ====================

def area_rectangulo():
    """
    Función para calcular el área de un rectángulo
    Fórmula: A = b × h
    """
    print("\n" + "🟦" * 50)
    print("🟦              ÁREA DEL RECTÁNGULO                     🟦")
    print("🟦" * 50)
    print("\n📐 Fórmula: A = base × altura")
    print("-" * 50)

    # Solicitar la base
    base = float(input("\n📏 Ingresa la base del rectángulo: "))

    # Solicitar la altura
    altura = float(input("📏 Ingresa la altura del rectángulo: "))

    # Calcular el área
    area = base * altura

    # Mostrar resultados
    print("\n" + "✨" * 50)
    print("📊 RESULTADOS:")
    print(f"   • Base ingresada: {base}")
    print(f"   • Altura ingresada: {altura}")
    print(f"   • Fórmula aplicada: A = {base} × {altura}")
    print(f"   • 🎯 Área del rectángulo: {area:.2f} unidades²")
    print("✨" * 50)

def perimetro_rectangulo():
    """
    Función para calcular el perímetro de un rectángulo
    Fórmula: P = 2 × (b + h)
    """
    print("\n" + "🟩" * 50)
    print("🟩            PERÍMETRO DEL RECTÁNGULO                  🟩")
    print("🟩" * 50)
    print("\n📐 Fórmula: P = 2 × (base + altura)")
    print("-" * 50)

    # Solicitar la base
    base = float(input("\n📏 Ingresa la base del rectángulo: "))

    # Solicitar la altura
    altura = float(input("📏 Ingresa la altura del rectángulo: "))

    # Calcular el perímetro
    perimetro = 2 * (base + altura)

    # Mostrar resultados
    print("\n" + "✨" * 50)
    print("📊 RESULTADOS:")
    print(f"   • Base ingresada: {base}")
    print(f"   • Altura ingresada: {altura}")
    print(f"   • Fórmula aplicada: P = 2 × ({base} + {altura})")
    print(f"   • 🎯 Perímetro del rectángulo: {perimetro:.2f} unidades")
    print("✨" * 50)

def menu():
    """
    Función principal que muestra el menú y controla el programa
    Usa un ciclo while para mantener el programa activo
    """
    opcion = 0

    while opcion != 3:
        # Mostrar menú principal
        print("\n" + "🌟" * 50)
        print("🌟           MENÚ PRINCIPAL - RECTÁNGULO            🌟")
        print("🌟" * 50)
        print("\n           📐 CÁLCULOS GEOMÉTRICOS 📐")
        print("\n           1️⃣  Calcular área del rectángulo")
        print("           2️⃣  Calcular perímetro del rectángulo")
        print("           3️⃣  Salir del programa")
        print("\n" + "🌟" * 50)

        # Leer la opción del usuario
        opcion = int(input("\n💡 Selecciona una opción (1-3): "))

        # Estructura if-elif-else para manejar las opciones
        if opcion == 1:
            # Llamar a la función para calcular área
            area_rectangulo()
        elif opcion == 2:
            # Llamar a la función para calcular perímetro
            perimetro_rectangulo()
        elif opcion == 3:
            # Mensaje de despedida
            print("\n" + "🎉" * 50)
            print("🎉                                              🎉")
            print("🎉         ¡GRACIAS POR USAR EL PROGRAMA!      🎉")
            print("🎉              👋 ¡Hasta pronto! 👋            ")
            print("                                              🎉")
            print("🎉" * 50)
        else:
            # Mensaje de error para opción inválida
            print("\n" + "⚠️" * 50)
            print("⚠️                                              ️")
            print("⚠️            ⚠️ OPCIÓN NO VÁLIDA ⚠️             ⚠️")
            print("⚠️       Por favor, selecciona una opción del 1 al 3      ⚠️")
            print("⚠️                                              ⚠️")
            print("⚠️" * 50)

# ==================== PROGRAMA PRINCIPAL ====================

if __name__ == "__main__":
    # Mostrar la documentación inicial con fórmulas y explicaciones
    print(__doc__)

    # Pausa para que el usuario lea la información
    input("💾 Presiona ENTER para continuar al menú principal...")

    # Llamar a la función del menú para iniciar el programa
    menu()
```
